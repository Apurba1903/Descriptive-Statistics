### `Q.1`. Given a normal distribution with mean as 50 and deviation as 10, answer below questions

(X ~ N(5O, 10) ):

    a. what are the values of the mean and standard deviation?
    
    b. What value of x has a z-score of 1.4?

    c. What is the Z-score that corresponds to x = 30?

    d. What is the difference between positive and negative z values?

`Answer A`<br>
The mean is 50 and the standard deviation is 10.<br>


`Answer B`<br>
To find the value of x which corresponds to a z-score 1.4.<br>
We have to use the formula for converting the z-score to raw score (x):<br>
z = (x - μ) / σ<br>

Where,  μ = mean<br>
        σ = standard deviation<br>
        z = z-score<br>
        x = raw score<br>

After rearranging the formula, we get,<br>
x = z * σ + μ<br>
Where,  z = 1.4<br>
        μ = 50<br>
        σ = 10<br>

We get,<br>
x = 1.4 * 50 + 10 = 64<br>

Therefore, The value of x that has a z-score of 1.4 is 64<br>


`Answer C`<br>
To find the z-score that corresponds to x = 30, <br>
We use the same formula as that we used in Answer B,<br>
But we will solve for z now:<br>
z = (x - μ) / σ<br>

Where,  μ = mean<br>
        σ = standard deviation<br>
        z = z-score<br>
        x = raw score<br>
<br>
Values are, x = 30<br>
            μ = 50<br>
            σ = 10<br>

We get,<br>
z = (30 - 50) / 10 = -2<br>

Therefore, The value of z-score that has a x of 30 is -2<br>


`Answer D`<br>
 The sign of the z-score indicates whether the raw score (x) is above or below the mean of the distribution. A positive z-score indicates that the raw score is above the mean, while a negative z-score indicates that the raw score is below the mean. The absolute value of the z-score represents the number of standard deviations away from the mean.<br>

````python
# Code Here

import numpy as np
import matplotlib.pyplot as plt

# Given (X ~ N(5O, 10))
plt.hist(np.random.normal(500, 10, 10000), density=True, bins=50)
plt.show()

# a. Mean and Standard Deviation
mean = 50
std_dev = 10

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# b. Value of x with z-score 1.4
z_score = 1.4
x = z_score * std_dev + mean

print("The value of x with z-score of 1.4 is :", x)

# c. Z-score that corresponds to x = 30
x = 30
z_score = (x - mean) / std_dev

print("Z-score that corresponds to x = 30:", z_score)
````

### `Q.2:`The average test score in a certain statistics class was 74 with a standard deviation of 8. There are 2000 students in this class. Use the emperical rule to answer the following questions:

    (a) What percentage of students scored less than 58?

    (b) What is the probability that a student score between 66 and 82 on the exam?

    (c) How many students scored at most 90?

    (d) What percentage of students scored at least 66?

    (e) How many students scored more than 98 on the test?

`Given:`<br>
Mean Score, (μ) =  74<br>
Standard Deviation, (σ) = 8<br>
Number of Students, (N) = 2000<br>

The empirical rule states that for a normal distribution, approximately:<br>
- 68% of the data falls within one standard deviation of the mean (i.e., μ ± σ)<br>
- 95% of the data falls within two standard deviation of the mean (i.e., μ ± 2σ)<br>
- 99.7% of the data falls within three standard deviation of the mean (i.e., μ ± 3σ)<br>

Using the empirical rule, we can answer the following questions:<br>

`Answer A:`<br>
Find the z-score of 58<br>
z   = (x - μ) / σ<br>
    = (58 - 74) / 8<br>
    = -2<br>

The percentage of students who scored less than 58 is the same as the percentage of students who are more than 2 standard deviations below the mean, which is approximately:<br>
P(z < -2) = P(z > 2) = 0.0228<br>

Therefore, approximately 2.28% of students scored less than 58.<br>

`Answer B:`<br>
Find the z-score of 66 and 82<br>
z1  = (x - μ) / σ<br>
    = (66 - 74) / 8<br>
    = -1<br>

z2  = (x - μ) / σ<br>
    = (82 - 74) / 8<br>
    = 1<br>

The probability of a student scoring between 66 and 82 is the same as the probability of being within one standard deviation of the mean, which is approximately:<br>
P(-1 < z < 1) = 0.68<br>

Therefore, approximately 68% of the students scored between 66 and 82.<br>

`Answer C:`<br>
Find the z-score of 66 and 82<br>
z   = (x - μ) / σ<br>
    = (90 - 74) / 8<br>
    = 2<br>

The percentage of students who scored at most 90 is the same as the percentage of students who are less than or equal to 2 standard deviation above the mean, which is approximately:<br>
P(z <= 2) = 0.9772<br>

Therefore, approximately 1954 students scored at most 90 out of 2000 students.<br>

`Answer D:`<br>
Find the z-score of 66<br>
z   = (x - μ) / σ<br>
    = (66 - 74) / 8<br>
    = -1<br>

The percentage of students who scored at least 66 is the same as the percentage of students who are at most 1 standard deviation below the mean, which is approximately:<br>
P(z >= -1) = 0.8413<br>

Therefore, approximately 84.13% of the students scored at least 66.<br>

`Answer E:`<br>
Find the z-score of 98<br>
z   = (x - μ) / σ<br>
    = (98 - 74) / 8<br>
    = 3<br>

The percentage of students who scored more than 98 is the same as the percentage of students who are more than 3 standard deviation above the mean, which is approximately:<br>
P(z > 3) = 0.0013<br>

Therefore, approximately 1.3 students who scored more than 98 out of 2000 students.<br>
However, we can not have a fraction of student, we can say that either 1 or 2 students scored more than 98.<br>

````python
# Code Here

import numpy as np
import matplotlib.pyplot as plt

# Generate some normal distribution data
mu = 0
sigma = 1
data = np.random.normal(mu, sigma, size=10000)

# Compute the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Compute the percentages for each standard deviation
percentages = {
    1: round(((np.sum((data >= mean - std) & (data <= mean + std))) / len(data)) * 100, 2),
    2: round(((np.sum((data >= mean - 2*std) & (data <= mean + 2*std))) / len(data)) * 100, 2),
    3: round(((np.sum((data >= mean - 3*std) & (data <= mean + 3*std))) / len(data)) * 100, 2)
}


fig = plt.figure(figsize=(16, 9))

# Plot the histogram of the data
plt.hist(data, bins=60, density=True, alpha= 0.5)

# Plot the empirical rule lines with markers and percentages
for i in range(-3, 4):
    x = mean + i * std
    y = 1/(np.sqrt(2 * np.pi) * std) * np.exp(-(x - mean)**2 / (2 * std**2))
    
    plt.plot([x, x], [0, y], linestyle = '--', color='r', alpha = 0.7)
    plt.annotate(f"{i} std\n ({percentages.get(abs(i), 0)}%)", (x, y), xytext=(x - 0.1, y + 0.01), fontsize= 8, color='r')


# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('Frequency')
plt.title('Normal Distribution of Empirical Rule')

# Show the plot
plt.show()
````

### `Q-3:` Normally distributed IQ scores have a mean of 100 and a standard deviation of 15. Use the standard z-table to answer the following questions:

What is the probability of randomly selecting someone with an IQ score that is

    (a) less than 80?

    (b) greater than 136?

    (c) between 95 and 110?

    (d) What IQ score corresponds to the 90th percentile?
    
    (e) The middle 30% of IQs fall between what two values?

`Answer A:`<br>
Find the z-score of 80<br>
z   = (x - μ) / σ<br>
    = (80 - 100) / 15<br>
    = -1.33<br>

Using the z-table, the probability of a randomly selected person having an IQ score less than 80 is approximately 0.0912 or 9.12%.<br>

`Answer B:`<br>
Find the z-score of 136<br>
z   = (x - μ) / σ<br>
    = (136 - 100) / 15<br>
    = 2.40<br>

If we look at the z-table, it will give probability of IQ <= 136.<br>
We want probability of IQ > 136.<br>
So, We need to take -1<br>
P(IQ <= 136) = 1 - 0.9918 = 0.0082<br>

Therefore, the probability of a randomly selected person having an IQ score greater than 136 is approximately 0.0082 or 0.82%.<br>

`Answer C:`<br>
Find the z-score of 95 and 110<br>
z1  = (x - μ) / σ<br>
    = (95 - 100) / 15<br>
    = -0.33<br>

z2  = (x - μ) / σ<br>
    = (110 - 100) / 15<br>
    = 0.67<br>

Using z-table, the probability of a randomly selected person having an IQ score between 95 and 110 is approximately 0.3819 or 38.19%.<br>

`Answer D:`<br>
The 90th percentile corresponds to a z-score of 1.28 (from the z-table).<br>
We can use this z-score to find the corresponding ID score:<br>
IQ  = 1.28(15) + 100<br>
    = 119.2<br>

Therefore, the IQ score corresponding to the 90th percentile is approximately 119.2.<br>

`Answer E:`<br>
The middle 30% of IQs corresponds to the area between the 35th and 65th percentile (Since the total area under the curve is 100% and we want the middle 30%)<br>

The z-score corresponding to the 35th and 65th percentiles are approximately -0.39 and 0.39, respectively from the z-table.<br>

We can use this z-scores to find the corresponding IQ scores:<br>

IQ1 = -0.39(15) + 10 = 94.65<br>
IQ2 = 0.39(15) + 10 = 105.85<br>

Therefore, The middle 30% of IQs fall between approximately 94.65 and 105.85.<br>

````python
import numpy as np
import matplotlib.pyplot as plt

# Mean and Standard Deviation
mu = 100
sigma = 15

# Create data points for x-axis
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

# Calculate the corresponding y-values (PDF)
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Initialize the plot
fig, ax = plt.subplots(figsize=(10,6))

# Plot the normal distribution curve
ax.plot(x, y, color='black', linewidth=2)

# Shade the areas corresponding to 1, 2 and 3 standard deviations from the mean
ax.fill_between(x, y, where=(x > mu - sigma) & (x <= mu + sigma), alpha=0.3, color='blue')
ax.fill_between(x, y, where=(x > mu - 2*sigma) & (x <= mu + 2*sigma), alpha=0.3, color='orange')
ax.fill_between(x, y, where=(x > mu - 3*sigma) & (x <= mu + 3*sigma), alpha=0.3, color='green')

# Add a vertical line to indicate the mean
ax.axvline(x=mu, color='green', linestyle='--', linewidth=2)

# Set axis labels and title
ax.set_xlabel('IQ Scores')
ax.set_ylabel('Probability Density')
ax.set_title('Normal Distribution of IQ Scores')

# Display the plot
plt.show()
````

`Given:`
- Mean IQ Score (μ) = 100
- Standard Deviation (σ) = 15

````python
import pandas as pd
import numpy as np
import scipy.stats as st

# Define the range of Z-scores (0.0 to 3.4) for rows
z_range_p_ve = np.arange(0.0, 3.5, 0.1)

# Define the column values (0.00 to 0.09)
col = np.arange(0.00, 0.10, 0.01)

# Create the DataFrame
z_table = pd.DataFrame(index=z_range_p_ve, columns=col)

# Fill the table using norm.cdf
for i in z_range_p_ve:
    for j in col:
        z_value = round(i + j, 2)
        z_table.loc[i, j] = round(st.norm.cdf(z_value), 4)

print("Positive Z-SCORE Table")
z_table
````

````python
import pandas as pd
import numpy as np
import scipy.stats as st

# Define the range of z-scores to be included in the table
z_range_n_ve = np.arange(-3.4, 0.1, 0.1)         # -3.4 to 0.0
col = np.arange(0.0, 0.10, 0.01)                 # 0.00 to 0.09

# Create a pandas dataframe to store the Z-table
z_table = pd.DataFrame(index=z_range_n_ve, columns=col)

# Fill the DataFrame with corresponding cumulative probabilities
for i in z_range_n_ve:
    for j in col:
        z_value = round(i + j, 2)
        z_table.loc[i, j] = round(st.norm.cdf(z_value), 4)

# Display table
print("Negative Z-SCORE Table")
z_table
````

Consider this "wine" dataset which can be loaded using above code. It has 13 numerical features and a target column with 3 class: 0, 1, 2.


* **Which feature is the best predictor of the wine class in the wine dataset?** (Using pdf plot)

* **Which two combination of features from below listed features are best for wine classification?** (2D pdf plot)
    * Proline
    * Flavanoids
    * Color intensity
    * OD280/OD315 of diluted wines
    * Alcohol

````python
# Code Here

import pandas as pd
import seaborn as sns
from sklearn.datasets import load_wine

# Load The Wine Dataset
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Plot the violin plot for the alcohol feature
for column in data.feature_names:
    sns.kdeplot(x=column, hue="target", data=df)
    plt.show()
````

Based on PDF plots, Flavanoids column is the best feature. <br>
PDF of this feature for different classes are less overlapping.

````python
# Using Pairplot
import seaborn as sns
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt

# Load the Wine Dataset
data = load_wine(as_frame=True).frame
wine = data[['proline', 'flavanoids', 'color_intensity', 'od280/od315_of_diluted_wines', 'alcohol', 'target']]

# Define the Pairplot with 2D KDE Plots
g = sns.pairplot(wine, hue="target", diag_kind='kde', height=4, aspect=1.25)

# Add 2D KDE Plots to all other subplots
for ax in g.axes.flat:
    if ax.get_xlabel() != "" and ax.get_ylabel() != "":
        sns.kdeplot(data=wine, x=ax.get_xlabel(), y=ax.get_ylabel(), hue='target', ax=ax)
        ax.set_xlabel(ax.get_xlabel(), fontsize=10)
        ax.set_ylabel(ax.get_ylabel(), fontsize=10)
        ax.tick_params(labelsize = 8)


# Adjust the layout
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.05, right=0.95, hspace=0.4, wspace=0.4)

# Show the plot
plt.show()
````

Based on 2D KDE plot - proline and od280/od315_of_diluted_wines making best feature. <br>
3 Classes are less overlapped.
